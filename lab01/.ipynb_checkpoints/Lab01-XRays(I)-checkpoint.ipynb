## write a general pipeline

```python
def findpeak(filename, name, angle_, cutoff = 0.1, min_val = 1e-3, save = True, no_peak = False):
    #get data
    data = np.loadtxt('./data/'+filename,comments='#',delimiter='\t',skiprows=1)
    angle = data[:,0] * np.pi /180 /2 # convert to radian and divide by 2 to get theta
    intensity = data[:,1]

    
    #filter by gaussian
#     filtered_intensity = ndimage.gaussian_filter(intensity, sigma = 1)
    filtered_intensity = intensity

    
    #cutoff
    threshold = cutoff * np.max(filtered_intensity)
    filtered_intensity_cutoff = filtered_intensity.copy()
    filtered_intensity_cutoff[filtered_intensity < threshold] = 0
    
    #find peak
    widths = np.linspace(1, 20, num = 100)
    peaks = signal.find_peaks_cwt(filtered_intensity_cutoff, widths)
    
    peaks = peaks[filtered_intensity_cutoff[peaks] > min_val]
    find_width = signal.peak_widths(filtered_intensity, peaks, rel_height = 0.5)
    print(find_width)
    #plot
    plt.plot(angle, filtered_intensity, '.')
    
    if not no_peak:
        plt.vlines(angle[peaks], 0,filtered_intensity_cutoff[peaks], 'r', linestyles = 'dashed' )
    plt.xlabel(r'$\theta$(rad)')
    plt.ylabel('intensity(filtered)(cps)')
    

    plt.title(r'Diffraction of X-rays on {:s} at {:s} radian'.format(name, angle_))
    if save:
        plt.savefig('./figs/'+filename[:-4]+'.png', dpi = 300)
    plt.show()
    return angle[peaks]
```

## Analysis

### PART A: Calculate Wavelength of the X-Ray

from $2 d sin(\theta)  = n \lambda\$ , one can find wavelength of the X-Ray by 

$$ \lambda = \frac{2 d sin (\theta)}{n}$$

where $d_{NaCl} = 5.6402/2  = 2.8201$ Angstrom

```python
d_NaCl = 2.8201e-1 #nm

n_lambda = 2 * d_NaCl * np.sin(peak_A)

err = 2 * d_NaCl * np.cos(peak_A) * 0.01 /2#(0.05) * np.pi/180

n = np.array([1,1,2,2,3])

lamda =  n_lambda/n
err = err/n

print(lamda)
print(err)
plt.hlines(n_lambda, 0, 1, 'b')
cur_axes = plt.gca()
cur_axes.axes.get_xaxis().set_visible(False)
plt.ylabel(r'$\lambda/n$ (nm)')
plt.savefig('lambda.png', dpi = 300)
```

```python
mean_lambda_1 = np.mean(lamda[[0,2]])
err_lambda_1 = np.std(lamda[[0,2]])
err_res_1 = 1/3 * (np.sum(err[[0,2]]**2))**0.5

mean_lambda_2 = np.mean(lamda[[1,3,4]])
err_lambda_2 = np.std(lamda[[1,3,4]])/2**0.5
err_res_2 = 1/3 * (np.sum(err[[0,2]]**2))**0.5

print(u'lambda1 is {:f} \u00B1 {:f} nm'.format(mean_lambda_1,err_lambda_1 ))
print(u'lambda2 is {:f} \u00B1 {:f} nm'.format(mean_lambda_2,err_lambda_2 ))
print('error from resolution is', err_res_1, err_res_2)
```

### PART B: Calculate lattice constant of LiF

from $2 d sin(\theta)  = n \lambda\$ , one can find d by 

$$ d = \frac{n \lambda}{2 sin(\theta)}$$

where $\lambda = 0.144687 \pm 0.001768, 0.158427 \pm 0.001704$ nm


```python
n = np.array([1,1,2])
lamda = np.array([mean_lambda_1, mean_lambda_2, mean_lambda_2])
e_lamda= np.array([err_lambda_1, err_lambda_2, err_lambda_2])

d_LiF = n * lamda /(2 * np.sin(peak_B)) 
err = 1/np.tan(peak_B) * 0.01/2 #np.pi/180 *0.05

err = d_LiF * (err**2 + (e_lamda/lamda)**2)**0.5
err_std = np.std(d_LiF)/(len(d_LiF)-1)**0.5
err_res_LiF = (np.sum(err**2))**0.5 /len(err)
d_LiF_mean = np.mean(d_LiF)

print(u'd is {:f} \u00B1 {:f} nm'.format(d_LiF_mean,err_res_LiF ))
print('error from std is', err_std)
```

### PART C: Calculate lattice constant of NaCl at 45 degree

```python
n = np.array([1,1])
lamda = np.array([mean_lambda_1, mean_lambda_2])
e_lamda= np.array([err_lambda_1, err_lambda_2])

d_NaCl_45 = n * lamda /(2 * np.sin(peak_C)) 
err = 1/np.tan(peak_C) * 0.01 /2#np.pi/180 * 0.05
err = d_NaCl_45 * (err**2 + (e_lamda/lamda)**2)**0.5
err_std = np.std(d_NaCl_45)/(len(d_NaCl_45)-1)**0.5
err_res_NaCl = (np.sum(err**2))**0.5 /len(err)
d_NaCl_45_mean = np.mean(d_NaCl_45)

print(u'd is {:f} \u00B1 {:f} nm'.format(d_NaCl_45_mean,err_res_NaCl ))
print('error from std is', err_std)
print('d sqrt(2) is {:f} \u00B1 {:f} nm'.format(d_NaCl_45_mean * 2**0.5,err_res_NaCl* 2**0.5))
```

### PART D: Calculate lattice constant of LiF at 45 degree

```python
n = np.array([1,1])
lamda = np.array([mean_lambda_1, mean_lambda_2])
e_lamda= np.array([err_lambda_1, err_lambda_2])

d_LiF_45 = n * lamda /(2 * np.sin(peak_D)) 
err = 1/np.tan(peak_D) * 0.01 /2#np.pi/180 *0.05
err = d_LiF_45 * (err**2 + (e_lamda/lamda)**2)**0.5
err_std = np.std(d_LiF_45)/(len(d_LiF_45)-1)**0.5
err_res_LiF2 = (np.sum(err**2))**0.5 /len(err)
d_LiF_45_mean = np.mean(d_LiF_45)

print(u'd is {:f} \u00B1 {:f} nm'.format(d_LiF_45_mean,err_res_LiF2 ))
print('error from std is', err_std)
print('d sqrt(2) is {:f} \u00B1 {:f} nm'.format(d_LiF_45_mean * 2**0.5,err_res_LiF2* 2**0.5))
```

## Compare lattice constant

```python
real_lattic = [d_NaCl*2, 0.4035]
calculated_lattice = [np.nan, d_LiF_mean * 2]
err_oriented = [0, err_res_LiF * 2]
orinted_lattic = [d_NaCl_45_mean * 2 * 2**0.5, d_LiF_45_mean* 2 * 2**0.5]
err_45 = [err_res_NaCl* 2**0.5, err_res_LiF2* 2**0.5]
objects = ('NaCl', 'LiF')
X = np.arange(2)
fig = plt.figure()
# fig.set_size_inches(4, 3)
ax = fig.add_axes([-0.01,0,1.0,1.0])
ax.bar(X + 0.0, calculated_lattice, width = 0.2,color ='#2A80D5', label = 'calculated lattice constant')
ax.bar(X + 0.25, orinted_lattic, width = 0.2, color = '#F28228', label ='calculated lattice constant from oriented crystal')
ax.bar(X + 0.50, real_lattic, width = 0.2, color ='#FDDB69', label = 'referenced lattice constant' )
plt.xticks(X + 0.25, objects)
plt.ylabel('lattice constant (nm)')
plt.legend()
plt.title('Comparison of Lattice Constant From Different Approach')

# Create labels
real_label = ['{:.3f}'.format(real) for real in real_lattic]
calc_label = [r'{:.3f} $\pm$ {:.3f}'.format(real, err) for real, err in zip(calculated_lattice, err_oriented)] 
orient_label = [r'{:.3f} $\pm$ {:.3f}'.format(real, err) for real, err in zip(orinted_lattic, err_45)] 
# Text on the top of each barplot
for i in range(len(X)):
    plt.text(x = X[i]+0.45 , y = real_lattic[i]+0.01, s = real_label[i], size = 10)
    plt.text(x = X[i]+0.13 , y = real_lattic[i]+0.01, s = orient_label[i], size = 10)
    if i>=1:
        plt.text(x = X[i]- 0.15 , y = real_lattic[i]+0.01, s = calc_label[i], size = 10)
plt.ylim(0, 0.7)
# plt.subplots(constrained_layout=True)
# plt.gcf().subplots_adjust(bottom=0.2)
# Adjust the margins
plt.savefig('lattice.png', dpi = 300)
```
